```python
# Dependencies
import pandas as pd
import math
```

```python
# Assign desired file path to variable
file_path = "Resources/purchase_data.csv"

# Read csv file using pandas
purchase_df = pd.read_csv(file_path)

# Clean up data - delete any potential null entries
clean_purchase = purchase_df.dropna()
```

### Player Count

```python
### Create a DataFrame with rows corresponding to unique players only
unique_players = clean_purchase.drop_duplicates(subset='SN')

### Count the total number of players and store as a DataFrame
player_count = unique_players['SN'].count()
player_count_df = pd.DataFrame([player_count], columns=['Player Count'])

### Display player count
player_count_df
```

### Purchasing Analysis (Total)

```python
# Find the number of unique items
unique_items = len(clean_purchase['Item ID'].unique())

# Find the average purchase price
average_price = round(clean_purchase['Price'].mean(),2)

# Find the total number of purchases
total_purchases = clean_purchase['Purchase ID'].count()

# Find the total revenue
total_revenue = clean_purchase['Price'].sum()
```

```python
# Assemple relevant data in a dictionary
purchasing_total_data = {'Number of Unique Items': unique_items,
                         'Average Purchase Price': average_price,
                         'Total Number of Purchases': total_purchases,
                         'Total Revenue': total_revenue,
                        }

# Build the DataFrame
purchasing_total_df = pd.DataFrame(purchasing_total_data, index=[0])

# Format the columns that include 'price' data
purchasing_total_df[['Average Purchase Price', 'Total Revenue']] = \
    purchasing_total_df[['Average Purchase Price', 'Total Revenue']].applymap('${:,.2f}'.format)

purchasing_total_df
```

### Gender Demographics

```python
# Group clean_purchase based on 'Gender'
gender_grouped = clean_purchase.groupby('Gender')
```

```python
# Find the unique players per gender
gender_unique_players = gender_grouped['SN'].unique()

# Count the number of unique players per gender
gender_player_count = [len(gender_unique_players[i]) for i in range(0, gender_unique_players.count())]

# Find the percentage of players per gender
gender_player_percent = (gender_player_count / player_count) * 100
```

```python
# Assemple relevant data in a dictionary
gender_demo_data = {'Player Count': gender_player_count,
                    'Percentage of Players': gender_player_percent, 
                   }

# Build the DataFrame
gender_demo_df = pd.DataFrame(gender_demo_data, index=['Female', 'Male', 'Other / Non-Disclosed'])

# Format the 'Percentage of Players' column
gender_demo_df['Percentage of Players'] = \
    gender_demo_df['Percentage of Players'].map('{:.2f}%'.format)

gender_demo_df
```

### Purchasing Analysis (Gender)

```python
# Find purchase count for each gender
gender_purchase_count = gender_grouped['Purchase ID'].count()

# Find average purchase price for each gender
gender_avg_price = gender_grouped['Price'].mean()

# Find the total purchase value for each gender
gender_purchase_value = gender_grouped['Price'].sum()

# Find the average purchase total per person by gender
gender_avg_total = gender_purchase_value / gender_player_count
```

```python
# Assemple relevant data in a dictionary
purchasing_gender_data = {'Purchase Count': gender_purchase_count,
                          'Average Purchase Price': gender_avg_price,
                          'Total Purchase Value': gender_purchase_value,
                          'Avg. Purchase Total Per Person': gender_avg_total,
                         }

# Build the DataFrame
purchasing_gender_df = pd.DataFrame(purchasing_gender_data, \
                                    index=['Female', 'Male', 'Other / Non-Disclosed'])

# Format the columns with 'price' data
purchasing_gender_df[['Average Purchase Price', 'Total Purchase Value', 'Avg. Purchase Total Per Person']] = \
    purchasing_gender_df[['Average Purchase Price', 'Total Purchase Value', 'Avg. Purchase Total Per Person']] \
        .applymap('${:,.2f}'.format)

purchasing_gender_df
```

### Age Demographics

```python
# Create a sequence of age brackets with bins of 4 years
bins = [4]
count = 0
while bins[count] < 39:
    bins.append(bins[count] + 5)
    count += 1

# The last bin will span 5 years and account for the 40+ age group
bins.append(bins[count] + 6)
```

```python
# Add a new column labeled 'Age Group' to clean_purchase
clean_purchase['Age Group'] = pd.cut(clean_purchase['Age'], bins, labels=False)

# Group clean_purchase based on 'Age Bracket'
age_grouped = clean_purchase.groupby('Age Group')
```

```python
# Find the unique players in every age group
age_unique_players = age_grouped['SN'].unique()

# Count the number of unique players in each group
age_player_count = [len(age_unique_players[i]) for i in range(0, age_unique_players.count())]

# Find the percentage of players in each age group
age_player_percent = (age_player_count/player_count) * 100
```

```python
# Assemple relevant data in a dictionary
age_demo_data = {'Player Count': age_player_count,
                 'Percentage of Players': age_player_percent,
                }

# Create index labels
age_idx = [f"<{10}"]
for i in range(10, 39, 5):
    age_idx.append(f"{i}-{i+4}")
age_idx.append(f"{40}+")

# Build the DataFrame
age_demo_df = pd.DataFrame(age_demo_data, index=age_idx)

# Format the 'Percentage of Players' column
age_demo_df['Percentage of Players'] = \
    age_demo_df['Percentage of Players'].map('{:.2f}%'.format)

age_demo_df
```

### Purchasing Analysis (Age)

```python
# Find the purchase count for each age group
age_purchase_count = age_grouped['Purchase ID'].count()

# Find the average purchase price for each age group
age_avg_price = age_grouped['Price'].mean()

# Find the total purchase value for each age group
age_purchase_value = age_grouped['Price'].sum()

# Find the average purchase total per person per age group
age_avg_total = age_purchase_value / age_player_count
```

```python
# Assemple relevant data in a dictionary
purchasing_age_data = {'Purchase Count': age_purchase_count,
                       'Average Purchase Price': age_avg_price,
                       'Total Purchase Value': age_purchase_value,
                       'Avg. Purchase Total Per Person': age_avg_total,
                      }

# Build the DataFrame
purchasing_age_df = pd.DataFrame(purchasing_age_data)

# Format the columns with 'price' data
purchasing_age_df[['Average Purchase Price', 'Total Purchase Value', 'Avg. Purchase Total Per Person']] = \
    purchasing_age_df[['Average Purchase Price', 'Total Purchase Value', 'Avg. Purchase Total Per Person']] \
        .applymap('${:,.2f}'.format)

purchasing_age_df.set_index([age_idx])
```

### Top Spenders

```python
# Group clean_purchase by 'SN'
players_grouped = clean_purchase.groupby('SN')

# Find the purchase count for every player
players_purchase_count = players_grouped['Purchase ID'].count()

# Find the total purchase price for every player
players_purchase_value = players_grouped['Price'].sum()

# Sort players in descending order of total purchase price
sorted_players = players_purchase_value.sort_values(ascending=False)
```

```python
# Pick out the top 5 spenders
top5_players = sorted_players[0:5]

top5_SN = [top5_players.index[i] for i in range(0,5)]
top5_purchase_count = [players_purchase_count[top5_players.index[i]] \
                       for i in range(0,5)]
top5_purchase_value = [players_purchase_value[top5_players.index[i]] \
                       for i in range(0,5)]
top5_avg_price = [(top5_purchase_value[i] / top5_purchase_count[i]) \
                  for i in range(0,5)]
```

```python
# DISPLAY RESULTS IN A FORMATTED DATAFRAME
### Assemple relevant data in a dictionary
top5_players_data = {'SN': top5_SN,
                     'Purchase Count': top5_purchase_count,
                     'Average Purchase Price': top5_avg_price,
                     'Total Purchase Value': top5_purchase_value,
                    }

### Build the DataFrame
top5_players_df = pd.DataFrame(top5_players_data)

### Format the columns with 'price' data
top5_players_df[['Average Purchase Price', 'Total Purchase Value']] = \
    top5_players_df[['Average Purchase Price', 'Total Purchase Value']].applymap('${:,.2f}'.format)

top5_players_df.set_index('SN')
```

### Most Popular Items

```python
# Group clean_purchase by 'Purchase ID'
itemID_grouped = clean_purchase.groupby('Item ID')

# Store item names
item_names = itemID_grouped['Item Name'].unique()

# Find the purchase count for every item
item_purchase_count = itemID_grouped['Purchase ID'].count()

# Find the total purchase price for every item
item_purchase_value = itemID_grouped['Price'].sum()

# Sort items in descending order of purchase count
sorted_items = item_purchase_count.sort_values(ascending=False)
```

```python
# Pick out the top 5 most popular games
top5_items = sorted_items[0:5]

top5items_ID = [top5_items.index[i] for i in range(0,5)]
top5items_name = [item_names[top5_items.index[i]][0] \
                  for i in range(0,5)]
top5items_pcount = [item_purchase_count[top5_items.index[i]] \
                    for i in range(0,5)]
top5items_pvalue = [item_purchase_value[top5_items.index[i]] \
                    for i in range(0,5)]
top5items_price = [(top5items_pvalue[i]/top5items_pcount[i]) \
                   for i in range(0,5)]
```

```python
# Assemple relevant data in a dictionary
top5_items_data = {'Item ID': top5items_ID,
                   'Item Name': top5items_name,
                   'Purchase Count': top5items_pcount,
                   'Item Price': top5items_price,
                   'Total Purchase Value': top5items_pvalue,
                  }

# Build the DataFrame
top5_items_df = pd.DataFrame(top5_items_data)

# Format the columns with 'price' data
top5_items_df[['Item Price', 'Total Purchase Value']] = \
    top5_items_df[['Item Price', 'Total Purchase Value']].applymap('${:,.2f}'.format)

top5_items_df.set_index('Item ID')
```
